```python
df.describe()
```

```python
num_features = ["pclass", "age", "sibsp", "parch", "fare"]
for f in num_features:
  plt.figure(figsize=(6, 4))
  plt.hist(df[f], bins=10, edgecolor='black')
  plt.title(f"{f} column: ")
```

```python
df['embarked'] = df['embarked'].astype(str)
categ_features = ["sex", "embarked"]
for f in categ_features:
    plt.figure(figsize=(6, 4))
    plt.hist(df[f], bins=3, edgecolor='black')
```

```python
columns_to_plot = ["pclass", "age", "sibsp", "parch", "fare"]
for i in columns_to_plot:
    df[[i]].plot(kind='box', vert=False, figsize=(14, 6))
    plt.title(f"{i} feature")
    plt.show()
```

```python
df_t = df.drop(["sex", "embarked"], axis="columns")
corr = df_t.corr(method='pearson')
corr
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr, cmap='RdBu', fignum=fig.number)

plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)

plt.colorbar()
plt.clim(-1, 1)
```

З навединих даних можна зробити висновок, що існує помірна негативна кореляція між класом пасажира і вартістю квитка (pclass та fare). Пасажири вищого класу (1-й клас) мають тенденцію платити більше за квитки, ніж пасажири нижчих класів (3-й клас). Також існує помірна позитивна кореляція між кількістю братів/сестер або чоловіка/дружини на борту і кількістю батьків/дітей на борту (sibsp та parch). Пасажири, які подорожують з родичами, мають тенденцію подорожувати з більшою кількістю членів сім'ї.
